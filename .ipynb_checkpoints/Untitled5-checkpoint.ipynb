```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 24.0, 0.1)
s = np.sin((t-13)/4)

plt.plot(t, s)
plt.text(0, -1, r'Moon inclinaison', fontsize=15)
plt.title(r'$\mathcal{A}\sin(\omega t)$', fontsize=20)
plt.xlabel('Time [h]')
plt.ylabel('Angle [degrees]')
plt.grid(True, linestyle='-.')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 24.0, 0.1)
t = np.arange(0, 90, 1)
s = np.sin(t/14)

plt.plot(t, s)
plt.text(0, -1, r'Moon inclinaison', fontsize=15)
plt.title(r'$\mathcal{A}\sin(\omega t)$', fontsize=20)
plt.xlabel('Time [s]')
plt.ylabel('Angle [degrees]')
plt.grid(True, linestyle='-.')
plt.show
```
